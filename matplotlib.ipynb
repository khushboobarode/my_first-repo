# matplotlib-Pyplot

```python
#plot
import matplotlib.pyplot as plt
plt.plot([1,2,3,4],[3,4,5,6],'r')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
#plot
import matplotlib.pyplot as plt
plt.plot([1,2,3,4],[3,4,5,6],'rs')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
#plot
import matplotlib.pyplot as plt
plt.plot([1,2,3,4],[3,4,5,6],'rs')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.axis([0,4,0,8])
plt.show()
```

```python
import numpy as np
t=np.arange(0,5,0.2)
plt.plot(t,t,"y--",t,t*2,"g^",t,t**2,"b*")
plt.show()
```
